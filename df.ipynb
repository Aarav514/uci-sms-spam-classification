```python
df['text'] = df['text'].str.lower()
df['text'] = df['text'].str.strip()
df
```

```python
# checking html tags
df['text'][df['text'].str.contains(r'<.*?>')]  # no html tag found
```

```python
# Regular expression pattern for removing urls and emails
import re
def removing_urls_emails(text):
     pattern = re.compile(r'experiencehttp://www.vouch4me.com/etlp/dining.asp\'')
     pattern1 = re.compile(r'((http)|(www)).*?(com|uk|net)([\S]+)?')
     pattern2 = re.compile(r'\swww.*?[\S]+')
     pattern3 = re.compile(r'\s\w+\.com(uk)?')
     pattern4 = re.compile(r'\s\w+\.co\.uk')
     pattern5 = re.compile(r'https?://(?:\S+\s*)+?(?:com|co\.uk|tv|net|ac)(?:/\S*)?')
     pattern6 = re.compile(r'info@[\S]+')
     pattern7 = re.compile(r'[\S]+\.co\.uk')
     pattern8 = re.compile(r'[\S]+\.com')
     text = re.sub(pattern,'',text)
     text = re.sub(pattern1,'',text)
     text = re.sub(pattern2,'',text)
     text = re.sub(pattern3,'',text)
     text = re.sub(pattern4,'',text)
     text = re.sub(pattern5,'',text)
     text = re.sub(pattern6,'',text)
     text = re.sub(pattern7,'',text)
     text = re.sub(pattern8,'',text)
     return text
     
    
          
    
df['text'] = df['text'].apply(removing_urls_emails)
```

```python
# Remove stopwords 
import nltk
from nltk.corpus import stopwords

def removing_stopwords(text):
    l = []
    stopword = stopwords.words('english')

    text = text.split()
    
    for i in text:
        if i not in stopword:
            l.append(i)
    return " ".join(l)

        

df['text'] = df['text'].apply(removing_stopwords)    
```

```python
# Remove special char + punctuations
def remove_punctuation_specialchar(text):
    pattern = re.compile(r'[^a-zA-Z0-9\s]')
    text = re.sub(pattern,'',text)
    return text.strip()
        


df['text'] = df['text'].apply(remove_punctuation_specialchar)  
```

```python
# No emoji found 
import emoji    
emoji.EMOJI_DATA
for i in df['text']:
    for ch in i:
        if ch in emoji.EMOJI_DATA:
            print(ch)
```
